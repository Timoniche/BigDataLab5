```python
import pathlib

from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .master("local") \
    .appName("KMean") \
    .getOrCreate()
```

```python
def load_dataset(dataset_path):
    dataset = spark.read.csv(dataset_path, header=True, inferSchema=True)
    dataset.show(5)
```

```python
curdir = str(pathlib.Path().parent.resolve())
dataset_filename = 'small.csv'
dataset_path = curdir + '/' + dataset_filename

load_dataset(dataset_path)
```
